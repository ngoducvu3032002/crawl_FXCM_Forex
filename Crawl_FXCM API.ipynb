```python
list_currency= ['EUR/USD',  'USD/JPY',  'GBP/USD', 'USD/CHF', 'EUR/CHF','AUD/USD',
 'USD/CAD','NZD/USD', 'EUR/GBP','EUR/JPY', 'GBP/JPY','CHF/JPY', 'GBP/CHF','EUR/AUD',
  'EUR/CAD','AUD/CAD', 'AUD/JPY','CAD/JPY', 'NZD/JPY','GBP/CAD', 'GBP/NZD','GBP/AUD', 
  'AUD/NZD','USD/SEK', 'EUR/SEK','EUR/NOK', 'USD/NOK','USD/MXN', 'AUD/CHF','EUR/NZD',]
```

```python
print(list_currency)
```

```python
name='EUR/USD'
```

```python
#Khai báo thời gian
import datetime as dt
# start = dt.datetime(2021, 1, 1)
# stop = dt.datetime(2022, 7, 1)
start = dt.datetime(2022, 1, 1)
stop = dt.datetime(2022, 7, 1)

```

```python
full=pd.DataFrame()
for currency in list_currency:
    temp=currency[:3]+currency[4:7]
    df2=con.get_candles(currency, period='H1',start=start, stop=stop)
    df2['tic']=temp
    full= pd.concat([full, df2])
```

```python
full
```

```python
df=full
df['adjcp']=df[['bidclose','askclose']].mean(axis=1)
df['open']=df[['bidopen','askopen']].mean(axis=1)
df['high']=df[['bidhigh','askhigh']].mean(axis=1)
df['low']=df[['bidlow','asklow']].mean(axis=1)
```

```python
df
```

```python
dataset = df[['tic','adjcp', 'open', 'high', 'low', 'tickqty']]
dataset.columns = ['tic','close', 'open', 'high', 'low', 'volume']
```
